```python
#Abstraction


from abc import ABC, abstractmethod

class BankApp(ABC):

  def database(self):
    print("Connected to database")

  @abstractmethod #abb chile class ko zhakk mar ke security methode ko use karna hoga
  def security(self):
    pass
```

```python
class MobileApp(BankApp):

  def mobile_login(self):
    print("Login into mobile")

  def security(self): #agar ise nhi likha to object nhi banega bcz this is abstract methode
    print("Mobile is secure")
```

```python
m1 = MobileApp()
m1.mobile_login()
```
