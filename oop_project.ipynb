```python
import nlpcloud

class NLPApp:
  def __init__(self):
   self. __database  = {}
   self.__first_menu()

  def __first_menu(self):
    first_input = input("""
    Hi ! How would you like to proceed ?
    1. Not a member ? Register
    2. Already a member? Login
    3. Galti se aa gaye? Exit
    """)

    if first_input == '1':
      self.__register()
    elif first_input == '2':
      self.__login()
    else:
      exit()

  def __register(self):
    input_name  =  input("Enter Your Name: ")
    input_email = input("Enter Your Email: ")
    input_PassWord = input("Enter Your PassWrod: ")

    if input_email in self.__database:
      print("User Is Already Exist ! Please Login...")
      self.__login()
    else:
      self.__database[input_email] = [input_name, input_PassWord] # Corrected: Assign directly to the key
      print("Registration Is Successful, Please Login..")
      self.__login()

  def __login(self):
    email  = input("Enter Your Registered email: ")
    PassWord = input("Enter Your PassWord: ")

    if email in self.__database and self.__database[email][1] == PassWord: # Added password check
      print("Login Succefully.. !")
      self.__second_menu()
    else:
      print("Wrong Email or PassWord !! Please Try Again...") # Updated error message
      self.__login()

  def __second_menu(self):
    second_input = input("""
    Hi ! How would you like to proceed ?
    1. NER
    2. Language Detection
    3. Sentiment Analysis
    4. Logout
    """)

    if second_input == '1':
      self.__NER()
    elif second_input == '2':
      self.__langDetection()
    elif second_input == '3':
      self.__SentiAnalysis()
    else:
      exit()

  def __NER(self):

    para = input("Enter The Paragraph: ")
    search_term = input("Enter what would you like to search? : ")
    client = nlpcloud.Client("gpt-oss-120b", "2e69f7f4e012a214cb723f0bcc300454d1ec9942", gpu=True)
    response = client.entities( para,  searched_entity=search_term )
    print(response)

  def __langDetection(self):

    StateMent = input("Enter Your Statement in you'r language: ")
    client = nlpcloud.Client("python-langdetect", "2e69f7f4e012a214cb723f0bcc300454d1ec9942", gpu=False)
    Response = client.lang_detection(StateMent)
    print(Response)

  def __SentiAnalysis(self):

    StateMent = input("Enter You'r Emotional Statement: ")
    client = nlpcloud.Client("gpt-oss-120b", "2e69f7f4e012a214cb723f0bcc300454d1ec9942", gpu=True)
    Response = client.sentiment(StateMent,target="NLP Cloud")
    print(Response)

```
